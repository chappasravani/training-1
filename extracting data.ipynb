```python
import PyPDF2, urllib, nltk
from io import BytesIO
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
wFile=urllib.request.urlopen('file:///C:/Users/chapp/AppData/Local/Microsoft/Windows/INetCache/IE/YY8L8ZTW/file1pdf[1].pdf')
pdfreader=PyPDF2.PdfReader(BytesIO(wFile.read()))
```

```python
pageObj=pdfreader.pages[2]
page5=pageObj.extract_text()
punctuations=['(',')',';',':','[',']',',','...','.']
tokens=word_tokenize(page5)
stop_words=stopwords.words('english')
keywords=[word for word in tokens if not word in stop_words not in punctuations]
```

```python
keywords
```

```python
name_list=list()
check=['Mr.','Mrs.','Ms.']
for idx, token in enumerate(tokens):
    if token.startswith(tuple(check)) and idx <(len(tokens)-1):
        name=token+tokens[idx+1]+' '+tokens[idx+2]
        name_list.append(name)
print(name_list)        
        
```
